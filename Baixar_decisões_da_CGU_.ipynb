```python
import requests
from bs4 import BeautifulSoup
import os
import re
from google.colab import files
from zipfile import ZipFile
```

```python
# Função para baixar um arquivo PDF
def download_pdf(url, filename):
    response = requests.get(url, stream=True)
    with open(filename, 'wb') as file:
        file.write(response.content)

# URL base da página
base_url = "https://buscaprecedentes.cgu.gov.br"

# URL da página com os resultados
search_url = base_url + "/?handler=search&ConsultaBasica.TermoPesquisa=CGU&ConsultaBasica.IdOuvidoriaSelecionada=&ConsultaBasica.OuvidoriaSelecionada=&ConsultaBasica.IdTipoDecisaoSelecionada=&ConsultaBasica.TipoDecisaoSelecionada=&ConsultaBasica.IdInstanciaSelecionada=&ConsultaBasica.InstanciaSelecionada=&numPagina=0&maximoRegistrosPorPagina=10000"

# Envie uma solicitação GET para a URL
response = requests.get(search_url)
html_content = response.content

# Analise o HTML para encontrar os links dos PDFs
soup = BeautifulSoup(html_content, 'html.parser')
pdf_links = soup.find_all('a', href=True)

# Filtre apenas os links de PDF
pdf_urls = [base_url + link['href'] for link in pdf_links if "idAnexo" in link['href']]
total_pdfs = len(pdf_urls)

# Crie um diretório para os PDFs
pdf_directory = 'pdf_downloads'
os.makedirs(pdf_directory, exist_ok=True)

# Baixe cada PDF e informe o progresso
for i, pdf_url in enumerate(pdf_urls, start=1):
    # Extrair o nome do arquivo do URL
    filename = re.findall(r"fileName=(.*?)&", pdf_url)
    filename = filename[0].replace('%20', ' ') if filename else f"documento_{i}.pdf"
    filepath = os.path.join(pdf_directory, filename)

    download_pdf(pdf_url, filepath)
    print(f"PDF escaneado: {i} de {total_pdfs} ({(i/total_pdfs)*100:.2f}% completo)")

# Compacte os PDFs em um arquivo zip
zip_filename = 'pdf_downloads.zip'
with ZipFile(zip_filename, 'w') as zipf:
    for root, dirs, files in os.walk(pdf_directory):
        for file in files:
            zipf.write(os.path.join(root, file))

# Inicie o download do arquivo zip para o computador do usuário
files.download(zip_filename)
```
